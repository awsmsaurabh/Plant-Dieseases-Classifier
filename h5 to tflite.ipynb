```
import os
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```
download = drive.CreateFile({'id': '1v_evL_zZTZb6sxFHLWO9KtMinFWFQFit'})
download.GetContentFile('PlantKeras-23-finetune_all-epoch-l1-001-drpt-04-lr0000001.h5')
```

```
import tensorflow as tf

converter = tf.lite.TFLiteConverter.from_keras_model_file("PlantKeras-23-finetune_all-epoch-l1-001-drpt-04-lr0000001.h5")
tflite_model = converter.convert()
open("PlantKeras-23-finetune_all-epoch-l1-001-drpt-04-lr0000001.tflite", "wb").write(tflite_model)
```

```
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```
upload = drive.CreateFile({'title': 'PlantKeras-23-finetune_all-epoch-l1-001-drpt-04-lr0000001.tflite'})
upload.SetContentFile('PlantKeras-23-finetune_all-epoch-l1-001-drpt-04-lr0000001.tflite')
upload.Upload()
```
